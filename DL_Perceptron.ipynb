```
#importing lib
import tensorflow as tf
import numpy as np
from keras.models import Sequential
from keras.layers import Dense
from tensorflow import keras
import matplotlib.pyplot as plt
import pandas as pd
```

```
from google.colab import drive
drive.mount('/content/drive')
```

```
re= pd.read_csv('/content/drive/MyDrive/datagdrive/Admission_Predict.csv')
```

```
X= re.iloc[:,-3].values #CGPA
Y= re.iloc[:,-1].values #Chance of admit
```

```
Y.dtype
```

```
x=X[:10]
y=Y[:10]
```

```
y
```

```
#defining a single layer; single neuron model
model = tf.keras.Sequential([keras.layers.Dense(units=3, activation='relu', input_shape=[1])])
model.add(Dense(1))
print(model.summary())
```

```
tf.keras.utils.plot_model(
    model,
    to_file="model.png",
    show_shapes=True,
    show_dtype=False,
    show_layer_names=True,
    rankdir="TB",
    expand_nested=False,
    dpi=96,
)
```

```
#defining the model evaluation for backpropogation
model.compile(optimizer='sgd', loss='mean_squared_error')
```

```
#taraining the model with iterations (No. of front & back propogation)
run_model = model.fit(x, y, epochs=800)
```

```
print(model.predict([8.0]))
```

66.9% which is very close to the real output which is close to 72%.

```
# list all data in history
print(run_model.history.keys())
```

```
# summarize history for accuracy
plt.plot(run_model.history['loss'])
plt.title('model loss')
plt.ylabel('loss')
plt.xlabel('epoch')
plt.show()
```
